# Pillow 를 이용한 Image Processing

## 연하장 만들기

- ``PIL(Python Image Library)`` 은 ``Pillow`` 의 old name  


    - 설치는 pip install pillow  
    - 사용은 import PIL

```python
import sys
import requests
from PIL import Image, ImageFilter, ImageDraw, ImageFont
from io import BytesIO
import base64

url = "https://dimg.donga.com/ugc/CDB/SHINDONGA/Article/5d/c2/67/30/5dc267300fefd2738de6.jpg"
```

```python
response = requests.get(url)
img = Image.open(BytesIO(response.content)).resize((350, 470))
img
```

```python
img.save('SeasonsGreeting2022.jpg', 'jpeg')
```

### 흑백 사진으로 변경

```python
grayscale = img.convert('L')    # L : black and white mode
grayscale
```

```python
# cropping (left, upper, right, lower) pixel coordinates
cropped = img.crop((100, 100, 350, 350))     
cropped
```

### Watermark 생성

- 한글 폰트가 필요할 경우 아래 링클를 통해 다운 받을 수 있습니다.

[네이버에서 제공하는 한글 글꼴 링크](https://hangeul.naver.com/font) 

- windows 한글 폰트 경로 `C:\Windows\Fonts`

```python
img = Image.open('SeasonsGreeting2022.jpg')
idraw = ImageDraw.Draw(img)
text = '홍길동님\n새해 복 많이 받으세요'

font = ImageFont.truetype("gulim.ttc", size=25)

idraw.text((70, 30), text, font=font, fill=(125,0,125))
img
```

```python
img.filter(ImageFilter.EMBOSS)
```

```python
img.filter(ImageFilter.CONTOUR)
```
