```python
import pandas as pd 

product_df = pd.DataFrame({"ProductID": [4109, 1412, 8931],
                           "Price": [5.0, 0.5, 1.5],
                           "Product": ["Sushi Roll", "Egg", "Bagel"]})

product_df = product_df.set_index('ProductID')

invoice_df = pd.DataFrame({"Customer": ["Ali", "Eric", "Ande", "Sam"],
                           "ProductID": [4109, 1412, 8931, 4109],
                           "Quantity": [1, 12, 6, 2]})

answer_df = pd.merge(invoice_df, product_df, right_index=True, left_on="ProductID")

print(product_df)
print()
print(invoice_df)
print()
print(answer_df)
```

## 연습 문제

- 위의 cars.csv 파일의 column 명이 너무 길므로 column 명 중 `(kW)` 를 `kw` 로,  `CITY (kWh/100 km)` 을 `city-km` 으로 변경  

- groupby 함수를 사용하여,  연도별 평균 배터리 용량(`kw`) 표시 (`Make` 무시)

```python
df = pd.read_csv("cars.csv")
df.head()
```

```python
df = df.rename(columns={'(kW)': 'kw', 'CITY (kWh/100 km)': 'city-km'})
df.head()
```

```python
df[['YEAR', 'kw']].groupby('YEAR').mean()
```
